# Part One

```python
def read_movement(file_name):
    with open(file_name, "r") as file:
        movement = []
        for row in file:
            direction, value = row[0], row[row.find(" ")+1:]
            value = int(value.split("\n")[0])
            movement.append((direction, value))
        return movement
```

```python
def calculate_final_depth(direction_with_value):
    horizontal = 0
    vertical = 0
    for movement in direction_with_value:
        if movement[0] == "f":
            horizontal += movement[1]
        elif movement[0] == "d":
            vertical += movement[1]
        else:
            vertical -= movement[1]
    return horizontal*vertical
```

```python
file_name = 'input.txt'
direction_with_value = read_movement(file_name)
final_depth = calculate_final_depth(direction_with_value)
final_depth
```

# Part Two

```python
def calculate_final_depth_with_aim(direction_with_value):
    horizontal = 0
    vertical = 0
    aim = 0
    for movement in direction_with_value:
        if movement[0] == "f":
            horizontal += movement[1]
            vertical = vertical + aim * movement[1]
        elif movement[0] == "d":
            aim += movement[1]
        else:
            aim -= movement[1]
    return horizontal*vertical
```

```python
file_name = 'input.txt'
direction_with_value = read_movement(file_name)
final_depth_with_aim = calculate_final_depth_with_aim(direction_with_value)
final_depth_with_aim
```
