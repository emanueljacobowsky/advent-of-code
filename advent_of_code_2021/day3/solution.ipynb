```python
def read_binary_code(file_name):
    with open(file_name, "r") as file:
        binary_code = []
        for row in file:
            binary_code.append(row.split("\n")[0])   
    return(binary_code)
```

```python
def calculate_power_consumption(transposed_binary_codes):
    gamma_rate = ""
    epsilon_rate = ""

    for column in transposed_binary_codes:
        if column.count(0) > column.count(1):
            gamma_rate += "0"
            epsilon_rate += "1"
        else:
            gamma_rate += "1"
            epsilon_rate += "0"

    return int(gamma_rate, 2) * int(epsilon_rate, 2)
```

```python
file_name = "input.txt"
binary_codes = read_binary_code(file_name)
transpose_binary_codes = transpose_binary_code(binary_codes)
power_consumption = calculate_power_consumption(transpose_binary_codes)
power_consumption
```

```python
def calculate_life_support(oxygen, co2):
    return int(oxygen[0], 2) * int(co2[0], 2)
```
