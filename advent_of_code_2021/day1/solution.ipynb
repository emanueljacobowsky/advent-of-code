# Part One

```python
def read_depths(file_name):
    with open(file_name, "r") as file:
        depths = []
        for row in file:
            depths.append(int(row.split("\n")[0]))
    return depths
```

```python
def count_depth_increasement(depths_list):
    depth_counter = 0
    for index, depth in enumerate(depths_list):
        if depths_list[index+1] > depth:
            depth_counter += 1
        if index+1 == len(depths_list)-1:
            return depth_counter
```

```python
file_name = 'C:\\Users\\Emanuel\\Programme\\VSC-workspace\\advent_of_code_2021\\day1\\input.txt'

depths_list = read_depths(file_name)
depth_increasement = count_depth_increasement(depths_list)

print(f"increasements: {depth_increasement}")
```

# Part Two

```python
def depth_pairs(depths_list):
    depths_list2 = depths_list.copy()
    depth_counter = 0
    for index in range(len(depths_list)):
        depth_pairs = zip(depths_list[:len(depths_list)-2], depths_list2[1:len(depths_list)-1], depths_list[2:])
    return depth_pairs
```

```python
def count_depth_sum_increasement(depth_pairs):
    depth_counter = 0
    depth_pairs = list(depth_pairs)
    for index, pair in enumerate(depth_pairs):
        if sum(depth_pairs[index+1]) > sum(pair):
            depth_counter += 1
        if index == len(depth_pairs)-2:
            return depth_counter
```

```python
file_name = 'C:\\Users\\Emanuel\\Programme\\VSC-workspace\\advent_of_code_2021\\day1\\input.txt'

depths_list = read_depths(file_name)
pairs_of_depth = depth_pairs(depths_list)
depth_sum_increasement = count_depth_sum_increasement(pairs_of_depth)

print(f"increasements: {depth_sum_increasement}")
```
